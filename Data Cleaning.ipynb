```python
def extract_job_title(resume_str):

    match = re.match(r"^[A-Z0-9(),.&/'\s-]+(?:[A-Z0-9(),.&/']{2,})", resume_str.strip())
    if match:
        return match.group(0).split("\n")[0].strip() 
    return None  

resume["job_title"] = resume["Resume_str"].apply(extract_job_title)

if resume["job_title"].isnull().any():
    print("Warning: Some resumes do not have a job title in uppercase at the beginning.")
```
